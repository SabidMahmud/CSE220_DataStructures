```python
class Patient:
    def __init__(self, id, name, age, blood_group):
        self.id = id
        self.name = name
        self.age = age
        self.blood_group = blood_group

class DublyNode:
    def __init__(self, elem, next = None, prv = None):
        self.data = elem
        self.next = next
        self.prev = prv

class WRM:
    def __init__(self):
        self.dhead = DublyNode(None, None, None) # Creating a dummy head
        self.tail = None

    def registerPatient(self, id, name, age, bg):
        print("\nExecuting registerPatient()")
        new_patient = Patient(id, name, age, bg)
        new_node = DublyNode(new_patient)

        if self.tail == None:
            self.dhead.next = new_node
            new_node.prev = self.dhead
            new_node.next = self.dhead
            self.dhead.prev = new_node
            self.tail = new_node
        else:
            self.tail.next = new_node
            new_node.prev = self.tail
            new_node.next = self.dhead
            self.dhead.prev = new_node
            self.tail = new_node
            # pass

    def servePatient(self):
        print("\nExeciting servePatient()")
        if self.dhead.next == None:
            print("No Patient in waiting room\n")
            return

        served = self.dhead.next
        self.dhead.next = self.dhead.next.next
        served.next.prev = self.dhead
        if served == self.tail:
            self.tail = self.dhead

        print(f"Serving patient: {served.data.name}\n")


    def showAllPatients(self):
        print("\nExecuting showAllPatients()")
        if self.dhead.next == None:
            print("No patients in the waiting room.\n")
            return

        print("Patients who are witing:")
        print("---------------------------")
        current_patient = self.dhead.next
        while current_patient != self.dhead:
            print(f"Patient ID {current_patient.data.id} . {current_patient.data.name}")
            current_patient = current_patient.next
        print("---------------------------")

    def canDocGoHome(self):
        print("\nExecuting canDocGoHome()")
        if self.dhead.next == None:
            return True
        return False

    def cancelAll(self):
        print("\nExecuting cancelAll()")
        if self.dhead.next == None:
            print("No appointments to cancel.\n")
            return
        self.dhead.next = None
        self.dhead.prev = None
        self.tail = None
        print("All appointments are canceled successfuly.\n")
```

```python
# Test code
wrm = WRM()
while True:
    print("***Waiting Room Management***")
    print("=============================\n")
    print("***Options:***")
    print("1. Add Patient")
    print("2. Serve Patient")
    print("3. Show All Patients")
    print("4. Can Doctor Go Home?")
    print("5. Cancel All Appointments")
    print("0. Exit")
    print("=============================\n")
    case = input("Enter your choice: ")
    print()

    if case == '0':
        break

    if case == '1':
        id = int(input("Enter patient ID: "))
        name = input("Enter patient's name: ")
        age = int(input("Enter patient's age: "))
        blood = input("Enter patient's blood group: ")
        wrm.registerPatient(id, name, age, blood)
        print("Patient added successfully!\n")

    elif case == '2':
        wrm.servePatient()

    elif case == '3':
        wrm.showAllPatients()

    elif case == '4':
        if wrm.canDocGoHome(): print("Doctor can go home now.\n")
        else: print("No, there are patients waiting.\n")

    elif case == '5':
        wrm.cancelAll()

    else:
        print("!!! Invalid choice. Please try again.\n")
```
