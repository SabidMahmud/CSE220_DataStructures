```python
#You must run this cell to print matrix and for the driver code to work
def print_matrix(m):
  row,col = m.shape
  for i in range(row):
    c = 1
    print('|', end='')
    for j in range(col):
      c += 1
      if(len(str(m[i][j])) == 1):
        print(' ',m[i][j], end = '  |')
        c += 6
      else:
        print(' ',m[i][j], end = ' |')
        c += 6
    print()
    print('-'*(c-col))
```

Task 1: Zigzag Walk

```python
def walk_zigzag(floor):
  row, col = floor.shape
  for i in range(row):
    if i%2 == 0 :
      for j in range(0,col,2):
        print(floor[i][j], end="" if j == row-1 or j == row-2 else " ")
    else:
      for j in range(col-1,0,-2):
        print(floor[i][j], end="" if j == 0 or j == 1 else " ")
    if i != row-1 : print()

floor = np.array([[ '3W' , '8B' , '4W' , '6B' , '1W' , '5B'],['3B' , '2W' , '1B' , '6W' , '3B' , '8W'],
                  ['9W' , '0B' , '7W' , '5B' , '3W' , '8B'],['2B' , '1W' , '3B' , '6W' , '0B' , '4W'],
                   ['1W' , '4B' , '2W' , '8B' , '6W' , '6B']])
print_matrix(floor)
print('Walking Sequence:')
walk_zigzag(floor)
#This should print
# 3W 4W 1W
# 8W 6W 2W
# 9W 7W 3W
# 4W 6W 1W
# 1W 2W 6W
```

Task 2: Landscape Screen

```python
def landscape(m):
  row, col = m.shape
  newArr=np.empty((col,row), dtype=int)
  for i in range(col):
    for j in range(row):
      newArr[i][j] = m[j][i]
  return newArr

m = np.random.randint(15,size=(4,3)) #This creates an np array of 4 x 3 where the values are random numbers from [0,15) range
print_matrix(m)
print()
print_matrix(landscape(m))
```

Task 3: Seating Arrangement

```python
# import numpy as np
def arrangement(m, diff):
  #To Do
  row,col = m.shape
  for i in range(row):
    for j in range(col):
      if i!=j and abs(m[i][j] - m[j][i]) != diff:
          return False
  return True

m = np.array([[7,13,9,14],[12,8,15,11],[10,17,3,2],[15,10,1,4]])
print_matrix(m)
print()
returned_value = arrangement(m,1)
print(returned_value)#This should print False
unittest.output_test(returned_value, False)

m = np.array([[8,15,7,12],[13,2,18,11],[9,20,5,2],[14,9,0,10]])
print_matrix(m)
print()
returned_value = arrangement(m,2)
print(returned_value)#This should print True
unittest.output_test(returned_value, True)
```

Task 4: Chess Piece

```python
def show_knight_move(knight):
  board = np.zeros((8,8), dtype = int)
  row, col = knight
  posibleMoves = np.array([(-2,-1), (-2, 1), (2,1), (2, -1), (-1, 2), (-1, -2), (1, 2), (1, -2)])
  board[row][col] = 33
  for mv in posibleMoves:
    i,j = mv
    if row+i >= 0 and row+i < 8 and col+j >= 0 and col+j < 8:
      board[row+i][col+j] = 3
  return board

knight = (3,4)
# knight = [int(i) for i in input().split()] # for custom inputs

chess_board = show_knight_move(knight)
print_matrix(chess_board)
```
