```python
#Run this cell
class Node:
  def __init__(self,elem,next = None):
    self.elem,self.next = elem,next

def createList(arr):
  head = Node(arr[0])
  tail = head
  for i in range(1,len(arr)):
    newNode = Node(arr[i])
    tail.next = newNode
    tail = newNode
  return head

def printLinkedList(head):
  temp = head
  while temp != None:
    if temp.next != None:
      print(temp.elem, end = '-->')
    else:
      print(temp.elem)
    temp = temp.next
  print()
```

Assemble Conga Line

```python
def conga_line(head):
  tail = head
  temp = head
  while tail != None:
    # print(temp.elem, tail.elem)
    if tail.elem < temp.elem:
      return False
    temp = temp.next
    tail = temp.next
  return True


head = createList(np.array([10,15,34,41,56,72]))
print('Original Conga Line: ', end = ' ')
printLinkedList(head)
returned_value = conga_line(head)
print(returned_value) #This should print True
unittest.output_test(returned_value, True)
print()

head = createList(np.array([10,15,44,41,56,72]))
print('Original Conga Line: ', end = ' ')
printLinkedList(head)
returned_value = conga_line(head)
print(returned_value) #This should print False
unittest.output_test(returned_value, False)
print()


```

Remove Compartment

```python
def countLen(head):
  l = 0
  t = head
  while t != None:
    l += 1
    t = t.next
  return l

def nodeAt(head, idx):
  count = 0
  temp = head
  obj = None
  while temp != None:
    if count == idx:
      obj = temp
      break
    temp = temp.next
    count+=1
  return obj

def remove_compartment(head,n):
  idx = countLen(head) - n
  if idx == 0:
    head = head.next
  elif idx >= 1 and idx < countLen(head):
    n1 = nodeAt(head, idx - 1)
    removed_node = n1.next
    n1.next =removed_node.next
    removed_node.next = None
  return head



head = createList(np.array([10,15,34,41,56,72]))
print('Original Compartment Sequence: ', end = ' ')
printLinkedList(head)
head = remove_compartment(head,2)
print('Changed Compartment Sequence: ', end = ' ')
printLinkedList(head) #This should print 10-->15-->34-->41-->72
print()

head = createList(np.array([10,15,34,41,56,72]))
print('Original Compartment Sequence: ', end = ' ')
printLinkedList(head)
head = remove_compartment(head,7)
print('Changed Compartment Sequence: ', end = ' ')
printLinkedList(head) #This should print 10-->15-->34-->41-->56-->72
print()

head = createList(np.array([10,15,34,41,56,72]))
print('Original Compartment Sequence: ', end = ' ')
printLinkedList(head)
head = remove_compartment(head,6)
print('Changed Compartment Sequence: ', end = ' ')
printLinkedList(head) #This should print 15-->34-->41-->56-->72
print()
```

Shuffle on index

```python
def shuffle_on_index(head):
  currentNode = head
  evenHead = currentNode
  evenTail = evenHead

  oddHead = currentNode.next
  oddTail = oddHead
  currentNode = oddTail.next

  i = 2
  while currentNode != None:
    if i%2 == 0:
      evenTail.next = currentNode
      evenTail = currentNode

    else:
      oddTail.next = currentNode
      oddTail = currentNode

    currentNode = currentNode.next

    if currentNode == None:
      oddTail.next = None

    i+=1
  evenTail.next = oddHead

  return evenHead

head = createList(np.array(['S','E','N','P','A','I']))
print('Original Play List: ', end = ' ')
printLinkedList(head)
newhead =  shuffle_on_index(head)
print('New Play List: ', end = ' ')
printLinkedList(newhead) #This should print S-->N-->A-->E-->P-->I
print()

head = createList(np.array(['N','I','S','H','I','N','O','Y','A']))
print('Original Play List: ', end = ' ')
printLinkedList(head)
newhead =  shuffle_on_index(head)
print('New Play List: ', end = ' ')
printLinkedList(newhead) #This should print N-->S-->I-->O-->A-->I-->H-->N-->Y
print()
```

Scavenger hunt

```python
def find_next_clue(head):
  currentNode = head
  while currentNode != None:
    currentMax = maximum = currentNode.elem
    maximum = currentNode.elem
    newNode = currentNode
    while newNode != None:
      if newNode.elem > currentMax :
        currentMax = newNode.elem
      newNode = newNode.next
    if maximum != currentMax:
      print(f"{currentMax} is greater than {maximum}")
    else:
      print(f"No next element greater than {maximum}")
    currentNode = currentNode.next




head = createList(np.array([7,85,54,16,11,30]))
print('Given Clue List: ', end = ' ')
printLinkedList(head)
find_next_clue(head)
print()

head = createList(np.array([20,13,33,12]))
print('Given Clue List: ', end = ' ')
printLinkedList(head)
find_next_clue(head)
print()
```
