```python
# Complete the functions defined in this cell

def rightRotate(source):
    temp = source[len(source)-1]
    for i in range(len(source)-1, 0, -1):
        source[i] = source[i-1]
    source[0] = temp
    return source

# Test 01: Play Right
def playRight(sequence,beats):
  for i in beats:
    if i == 1:
        rightRotate(sequence)
  return sequence


#Delete from array
def delete(arr, indx):
    for i in range(indx, len(arr)-1):
        arr[i] = arr[i+1]
    arr[len(arr)-1] = 0
    return arr
    # pass

# Test 02: Discard Cards
def discardCards(cards,number):
  i = 0
  while i < len(cards):
    if cards[i] == number:
        delete(cards, i)
    else:
        i+=1
  return cards


#Test 03: Merge Lineup
def mergeLineup(pokemon_1, pokemon_2):
    newArr = np.zeros(len(pokemon_1), dtype=int)
    for i in range(len(pokemon_1)):
        j = len(pokemon_1)-1-i
        if pokemon_1[i]==None:
            pokemon_1[i]=0
        if pokemon_2[j]==None:
            pokemon_2[j]=0

        newArr[i] = pokemon_1[i]+pokemon_2[j]
    return newArr


# Test 04: Balance your Salami
def balanceSalami(salami):
  sum = 0
  for i in salami:
    sum += i
  if sum%2 != 0: return False

  mySalami = 0
  for i in range(len(salami)):
    mySalami += salami[i]
    if sum - mySalami == mySalami: return True
  return False

# Test 05: Protecc Salami
def protectSalami(salami):
  for i in range(len(salami)):
    # flag = 0
    count = 1
    for j in range(i+1,len(salami)):
      if salami[j] == i:
        salami[i] = 0
        count += 1
    salami[i] = count
    count = 0
  
  
  for i in range(len(salami)):
    if salami[i] > 1:
      for j in range(i+1, len(salami)):
        if salami[i]==salami[j]: return True
  return False

# Test 06: Odd Even Wave
def waveYourFlag(arr):
  newArr = np.zeros(len(arr), dtype = int)
  temp = arr[0]
  if temp%2 == 0:
    evenid = 0
    oddid = 1
  else:
    evenid=1
    oddid=0
  for i in arr:
    if i%2 == 0:
      newArr[evenid] = i
      evenid += 2
    else:
      newArr[oddid] = i
      oddid += 2

  return newArr
```
