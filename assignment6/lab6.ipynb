### 1. [Very Easy]




```python
#(1a)
# find the factorial
def fact(n):
    if n == 0: return 1
    if n == 1: return 1
    return n * fact(n-1)

n = int(input("Input Number: "))
print(f"Factorial of {n} is {fact(n)}")
```

```python
#(1b) n-th Fibonacci
def findFibonacci(n):
    if n <= 1: return n
    else:
        return findFibonacci(n-1) + findFibonacci(n-1-1)

print(findFibonacci(int(input())))
```

```python
#(1c) Printing array

def printArr(arr, idx = 0):
    if len(arr) <= idx: return None
    print(arr[idx], end = ' ')
    printArr(arr, idx+1)

arr = [i for i in range(5)]

printArr(arr, 0)
```

```python
#(1d)
def powerN(num, pow):
    if pow == 0: return 1
    if pow == 1: return num
    return num * (powerN(num, pow-1))

num, pow = [int(i) for i in input().split()]
print(powerN(num, pow))
```

```python
#(2a) convert to binary

def binary_of(n):
    if n == 0:
        return 0
    else:
        return (n % 2) + (10 * (binary_of(int(n // 2))))


n = int(input())
print(f"Binary of {n} is: {binary_of(n)}")
```

```python
#(2b) adding linked list elements
import numpy as np
class Node:
    def __init__(self, data, next = None):
        self.data = data
        self.next = next

def createList(arr):
    head = Node(arr[0])
    tail = head
    for i in range(1, len(arr)):
        newNode = Node(arr[i])
        tail.next = newNode
        tail = newNode
    return head

array = np.array([1,4,5,6,3,4,7,89])
head = createList(array)

def count(head, add = 0):
    if head == None:
        return add
    add += head.data
    return count(head.next, add)

print( count(head) )
```

```python
#(2c) printing a linked list reversed
def reverseList(head):
    # pass
    if head == None:
        return
    reverseList(head.next)
    print(head.data, end = " ")

reverseList(head)
```

### 4. [Hard]

```python
#(4a)
def print_numberA(number):
    if number == 0:
        return
    else:
        print_numberA(number - 1)
        print(number, end = " ")

def patternA(sample_input, n):
    if n == sample_input + 1:
        return

    print_numberA(n)
    print()

    patternA(sample_input, n + 1)

sample_input = int(input())
patternA(sample_input, 1)
```

```python
#(4b)
def whitespace(num):
  if num == 0:
    return
  else:
    print(" ", end = " ")
    whitespace(num - 1)

def print_numberB(num):
  if num == 0:
    return
  else:
    print_numberB(num - 1)
    print(num, end = " ")


def patternB(x, n):
  if n == x + 1:
    return

  whitespace(x - n)
  print_numberB(n)
  print()

  patternB(x, n + 1)

x = int(input())
patternB(x, 1)
```
